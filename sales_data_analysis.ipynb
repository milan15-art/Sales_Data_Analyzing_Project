```python
                        

import pandas as pd
import numpy as np
```

```python
#  read csv data

df = pd.read_csv("Sales_data.csv")
# print(df)

# convert date column to Date time type
df["Date"] = pd.to_datetime(df["Date"])

# create month column

df["Month"]  = df["Date"].dt.to_period("M")

# calculate total sales per row

df["Total_sale"] = df["Quantity"]* df["Price"]

print("\nComplete sales Data")
print(df)


```

```python

# product wise sales 

product_sales = df.groupby("Product")["Total_sale"].sum()

print("\n ------Product wise Total Sales------")
print(product_sales)


# monthly sales using pivot table


monthly_sales = df.pivot_table(
                              values = "Total_sale",
                              index = "Month",
                              aggfunc = "sum")
print("\nMonthly Sale")
print(monthly_sales)
```

```python
# Monthly growth using numpy lib

sales_values = monthly_sales["Total_sale"].values
Monthly_Growth = np.diff(sales_values)


print("\n ---Monthly Growth---")
print(Monthly_Growth)
```

```python
# best and worst product sales

best_product_sales = product_sales.idxmax()
worst_product_sales = product_sales.idxmin()


print(f"Best product sales : {best_product_sales}")
print(f"Worst product sales : {worst_product_sales}")


```
